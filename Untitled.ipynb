```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import MinMaxScaler

# 1. Load your CSV file
df = pd.read_csv('upload_acf4d973-b455-4898-b450-b4127f96e4b9.csv')  # Replace with your CSV file name

# Optional: sort by date if not already sorted
df = df.sort_values('Date')

# 2. Use the 'Close' price for prediction
data = df['Close'].values.reshape(-1, 1)

# 3. Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# 4. Create features and labels for prediction
# We'll use the previous 10 days to predict the next day's price
def create_dataset(dataset, time_step=10):
    X, y = [], []
    for i in range(time_step, len(dataset)):
        X.append(dataset[i-time_step:i, 0])
        y.append(dataset[i, 0])
    return np.array(X), np.array(y)

time_step = 10
X, y = create_dataset(scaled_data, time_step)

# 5. Split into train and test sets
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# 6. Build and train the Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 7. Make predictions
y_pred = model.predict(X_test)
y_pred = scaler.inverse_transform(y_pred.reshape(-1, 1))
y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))

# 8. Plot the results
plt.figure(figsize=(12,6))
plt.plot(y_test_actual, label='Actual Price')
plt.plot(y_pred, label='Predicted Price')
plt.title('Stock Price Prediction (Random Forest)')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()

```
